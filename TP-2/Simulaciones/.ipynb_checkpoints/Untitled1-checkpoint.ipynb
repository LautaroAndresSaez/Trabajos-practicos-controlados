```python
import control
import matplotlib.pyplot as plt
import numpy as np
```

```python
z = control.TransferFunction.z
H = lambda w: 1 / ( np.exp( 1j*w ) - 1/2 )
```

```python
w = np.linspace( 0, np.pi, 2000 )
plt.polar( np.angle(H(w)), np.abs( H(w) ) )
```
